Prompt Engineering


````python
prompt_template = '''
请根据我提供的文本，帮我总结出文本中的知识，分成若干点，每一点应该完整独立。
不要额外加入文本中没有的内容或信息，你总结的每一个知识点都应该是在文本中表达出来的。
每一个知识点应该以 Json 的形式返回，其中包含以下四个键：
关键词：3~4个关键词，用于描述这个知识点的主要内容
应用场景：这个知识点在什么场景中适用
具体内容：知识点的具体内容，要详细具体，可以适当使用原文
注意事项；应用这个知识点时需要注意的事项
提供的文本：
```
{}
```
返回给我一个可解析的Python列表，每一项是上述四个键的字典。
'''
````

````python
prompt = prompt_template.format(s[:2000])
````

````python
prompt
````

````python

response = erniebot.ChatCompletion.create(
    model=model,
    messages=[{
        'role': 'user',
        'content': prompt
    }])
````

````python
print(prompt)
````

````python
print(response.rbody["result"])
````

批量处理

````python
from tqdm import tqdm

knowledges = []

for item in tqdm(files[:12]):
    if item.endswith(".txt"):
        with open(item, 'r') as f:
            s = f.read()
        start_loc = 0
        end_loc = 2000
        if len(s) < 2000:
            end_loc = len(s) - 1
        while start_loc < len(s) - 1:
            prompt = prompt_template.format(s[start_loc:end_loc])
            response = erniebot.ChatCompletion.create(
                model=model,
                messages=[{
                    'role': 'user',
                    'content': prompt
                }])
            print(response.rbody["result"])
            knowledges.append(response.rbody["result"])
            start_loc = end_loc
            end_loc = start_loc + 2000 if start_loc + 2000 < len(s) else len(s) - 1
````

````python
len(knowledges)
````

````python
js_lst = []
for item in knowledges:
    start_loc = item.find("{")
    end_loc = item.find("}")
    js_lst.append(eval(item[start_loc:end_loc + 1]))
````

````python
js_lst
````

````python
with open("knowledges.txt", "a") as f:
    for item in js_lst:
        for key in item:
            f.write(key + ": " + str(item[key]) + "\n")
        f.write("\n")
        f.write("###\n")
````
